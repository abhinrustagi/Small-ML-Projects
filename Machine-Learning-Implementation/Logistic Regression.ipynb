```python
import numpy as np
np.random.seed(0)
```

```python
X = np.array([[0, 1], [0, 0], [1, 0], [2, 0.5], [2, 1], [2, 3]])
Y = np.array([[0], [0], [0], [1], [1], [1]])
```

```python
weights = np.random.randn(3, 1)
print(weights)
```

```python
X = np.append(X, np.ones((6, 1)), axis = 1)
print(X)
```

```python
print(X.shape, Y.shape, weights.shape)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def costFunction(X, Y, weights):
    cost = sigmoid(X.dot(weights))
    cost = (-1/X.shape[0])*(Y.T.dot(np.log(cost))+(1-Y).T.dot(np.log(1-cost)))
    return cost
```

```python
def gradient(X, Y, weights):
    diff = (sigmoid(X.dot(weights)) - Y).T
    finalGradient = diff.dot(X)
    return finalGradient.T
```

```python
def gradientDescent(X, Y, learningRate, epochs):
    global weights
    for i in range(epochs):
        gradientCalc = gradient(X, Y, weights)
        weights = weights - (learningRate/X.shape[0])*gradientCalc
        print(costFunction(X, Y, weights))
    return weights
```

```python
newWeights = gradientDescent(X, Y, 0.1, 1000)
print(newWeights)
```
