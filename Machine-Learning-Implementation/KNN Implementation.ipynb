```python
def dist(x1, x2):
    return np.sqrt(sum((x1-x2)**2))

def KNN(X, Y, query_point, k=5):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(query_point, X[i])
        vals.append((d, Y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    print(vals)
    
    new_vals = np.unique(vals[:,1], return_counts = True)
    max_freq_index = new_vals[1].argsmax()
    
    return new_vals
```
