```python
import pandas as pd
import numpy as np
```

```python
def dist(x1, x2):
    return np.sqrt(sum((x1-x2)**2))

def KNN(X, Y, query_point, k=5):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(query_point, X[i])
        vals.append((d, Y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    print(vals)
    
    new_vals = np.unique(vals[:,1], return_counts = True)
    max_freq_index = new_vals[1].argmax()
    
    return max_freq_index
```

```python
x_train = pd.read_csv('Diabetes_XTrain.csv')
y_train = pd.read_csv('Diabetes_YTrain.csv')
```

```python
x_train.head()
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
x_test = pd.read_csv('Diabetes_XTest.csv')
```

```python
x_test = np.array(x_test)
```

```python
y_test = []
for x in range(len(x_test)):
    y_test.append(KNN(x_train, y_train, x_test[x]))
```

```python
with open("results.csv", "w") as f:
    f.write("Outcome\n")
    for i in y_test:
        f.write(str(i)+"\n")
```
